```python
FILE_PATH = r"C:\Users\Dell\Downloads\Amazon review data.html"

with open(FILE_PATH, "r", encoding="utf-8", errors="ignore") as f:
    soup = BeautifulSoup(f, "html.parser")

print("HTML file loaded successfully")
```

```python
texts = list(soup.stripped_strings)

print("Total text blocks found:", len(texts))
texts[:10]
```

```python
reviews = []

for t in texts:
    t = t.lower().strip()

    if len(t) >= 30:
        reviews.append(t)

print("Potential reviews extracted:", len(reviews))
```

```python
df = pd.DataFrame({"review_text": reviews})
df.head()
```

```python
def clean_text(text):
    text = re.sub(r"<.*?>", "", text) 
    text = re.sub(r"\s+", " ", text)    
    return text.strip()

df["review_text"] = df["review_text"].apply(clean_text)

df.drop_duplicates(inplace=True)

print("Final number of reviews:", len(df))
df.head()
```

```python
print("Dataset empty?", df.empty)

df.sample(5)
```

```python
df.to_csv("processed_reviews.csv", index=False)
print("Saved processed_reviews.csv successfully")
```
